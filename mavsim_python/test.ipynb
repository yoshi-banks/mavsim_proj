```python
import matplotlib.pyplot as plt
mylist = []
for i in range(1000000):
    my_val = np.random.normal(SENSOR.gyro_x_bias, SENSOR.gyro_sigma)
    mylist.append(my_val)
    
plt.hist(mylist, bins=100)
```
